```python
import geopandas as gpd

# Carregando o arquivo Shapefile da cidade de Belo Horizonte
shapefile_path = 'Mapa_BH\\REGIONAL.shp'
gdf = gpd.read_file(shapefile_path)

gdf.plot(cmap='jet', figsize=(10,10), )
```

```python
arq_regional = 'Mapa_BH\\REGIONAL.shp'
arq_bairros = 'Mapa_BH\\BAIRRO_OFICIAL.shp'
arq_hospitais = 'Mapa_BH\\HOSPITAIS\\HOSPITAIS.shp'
arq_favela = 'Mapa_BH\\VILA_FAVELA\\VILA_FAVELA.shp'

gdf_regional = gpd.read_file(arq_regional)
gdf_bairros = gpd.read_file(arq_bairros)
gdf_hospitais = gpd.read_file(arq_hospitais)
gdf_favela = gpd.read_file(arq_favela)

camada_bairro = gdf_bairros.plot(color='palegreen', edgecolor='green', figsize=(10,20))
camada_favela = gdf_favela.plot(ax=camada_bairro, color='yellow', edgecolor='orange', linewidth=1)
gdf_hospitais.plot(ax=camada_favela, marker='o', color='red', markersize=15)
```

```python
arq_populacao = 'Mapa_BH\\POP_DOMIC_BAIRRO_2010\\POP_DOMIC_BAIRRO_2010.shp'
gdf_populacao = gpd.read_file(arq_populacao)
gdf_populacao.plot( column='POPULACAO', edgecolor='grey', cmap='OrRd', legend=True,
                   legend_kwds={"label": "População de BH em 2010", "orientation": "horizontal"},
                   figsize=(10,15),
                   )
```
